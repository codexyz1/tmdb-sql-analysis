```python
# Connect to SQL Server
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "mssql+pyodbc://localhost\\SQLEXPRESS/TMDB?"
    "driver=ODBC+Driver+17+for+SQL+Server&"
    "trusted_connection=yes&"
    "TrustServerCertificate=yes"
)
```

```python
# Load Movies CSV into dbo.movies
movies = pd.read_csv("../data/raw/tmdb_5000_movies.csv")
movies.to_sql("movies", engine, if_exists="append", index=False)
print(f"Inserted {len(movies)} rows into movies")
```

```python
# Load Credits CSV into dbo.credits
credits = pd.read_csv("../data/raw/tmdb_5000_credits.csv")
credits.to_sql("credits", engine, if_exists="append", index=False)
print(f"Inserted {len(credits)} rows into credits")
```

```python
# Verify Row Counts
check_movies = pd.read_sql("SELECT COUNT(*) AS row_count FROM dbo.movies", engine)
check_credits = pd.read_sql("SELECT COUNT(*) AS row_count FROM dbo.credits", engine)
print("movies:", check_movies.iloc[0,0])
print("credits:", check_credits.iloc[0,0])
```
